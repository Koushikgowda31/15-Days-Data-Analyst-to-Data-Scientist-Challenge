```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('Indiranagar.csv')
```

```python
# Inspect the first few rows to understand structure
print(df.head())

# Clean column names if needed (remove leading/trailing spaces)
df.columns = df.columns.str.strip()

# Top 10 selling items by quantity
top_items = df.groupby('Item')['Qty.'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 6))
top_items.plot(kind='bar', color='green')
plt.title('Top 10 Best-Selling Items (Indiranagar Outlet)')
plt.ylabel('Quantity Sold')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Bottom 10 selling items by quantity
bottom_items = df.groupby('Item')['Qty.'].sum().sort_values(ascending=True).head(10)

plt.figure(figsize=(12, 6))
bottom_items.plot(kind='bar', color='red')
plt.title('Bottom 10 Least-Selling Items (Indiranagar Outlet)')
plt.ylabel('Quantity Sold')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv('Indiranagar.csv')
df.columns = df.columns.str.strip()

# Top 10 Selling Items
top_items = df.groupby('Item')['Qty.'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 6))
bars = plt.bar(top_items.index, top_items.values, color='green')
plt.title('Top 10 Best-Selling Items - Indiranagar Outlet')
plt.ylabel('Total Quantity Sold')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')

# Add value labels
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height + 1, f'{int(height)}',
             ha='center', va='bottom', fontsize=10, fontweight='bold')

plt.subplots_adjust(bottom=0.3)  # Give space for rotated x labels
plt.show()


# Bottom 10 Selling Items
bottom_items = df.groupby('Item')['Qty.'].sum().sort_values().head(10)

plt.figure(figsize=(12, 6))
bars = plt.bar(bottom_items.index, bottom_items.values, color='red')
plt.title('Bottom 10 Least-Selling Items - Indiranagar Outlet')
plt.ylabel('Total Quantity Sold')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')

# Add value labels
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height + 1, f'{int(height)}',
             ha='center', va='bottom', fontsize=10, fontweight='bold')

plt.subplots_adjust(bottom=0.3)  # Adjust bottom margin
plt.show()
```
