```python
# Load dataset

import pandas as pd

df =pd.read_csv("IMDB-Movie-Database.csv")
df.head()
```

```python
# Import Visualization libraries

import matplotlib.pyplot as plt
import seaborn as sns

# for better styling
sns.set_style("darkgrid")
```

```python
# Univariate visualizations

# 1. Top 10 Genre by Frequency

top_genres = df['Genre'].value_counts().head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top_genres.values, y=top_genres.index, hue=top_genres.index, palette='coolwarm')
plt.title("Top 10 Movie Genre")
plt.xlabel("Number of Movies")
plt.ylabel("Genre")
plt.show()
```

```python
# 2. Distribution of IMDB Ratings

plt.figure(figsize=(10,6))
sns.histplot(df['IMDb Score (1-10)'], bins=20, kde=True, color='skyblue')
plt.title("Distribution of IMDB Ratings")
plt.xlabel("IMDb Score (1-10)")
plt.ylabel("Count")
plt.show()
```

```python
# Bivariate visualization

# 1. Budget vs Gross Revenue

plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Budget', y='Gross Revenue', hue='Genre', palette='viridis')
plt.title("Budget vs Gross Revenue by Genre")
plt.xlabel("Budget (in millions)")
plt.ylabel("Gross Revenue (in millions)")
plt.legend(bbox_to_anchor=(1, 1))
plt.show()
```

```python
# 2. Average profitby Genre
df['Profit'] = df['Gross Revenue'] - df['Budget']
genre_profit = df.groupby('Genre')['Profit'].mean().sort_values(ascending=False).reset_index()

plt.figure(figsize=(12,6))
sns.barplot(x='Profit', y='Genre',hue='Genre', data=genre_profit.head(10), palette='magma')
plt.title("Top 10 Profitable Genres (by Average Profit)")
plt.xlabel("Average Profit")
plt.ylabel("Genre")
plt.show()
```

```python
# Correlation Heatmap

plt.figure(figsize=(12,8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Violin plot
top_Language = df['Language'].value_counts().head(10).index
filtered_df = df[df['Language'].isin(top_Language)]
plt.figure(figsize=(12, 6))


sns.violinplot(x='Language', y='IMDb Score (1-10)', data=filtered_df)
plt.title("IMDB Rating by Top 10 Language")
plt.xlabel("Language")
plt.ylabel("IMDb Score (1-10)")
plt.show()
```

```python
# Pie Chart

df['Genre'].value_counts().head(5).plot.pie(autopct='%1.1f%%')
plt.title("Top 5 Genres Distribution")
plt.ylabel("")
plt.show()
```

```python
# Scatter plot
sns.scatterplot(x='Budget', y='Gross Revenue', data=df)
plt.title("Budget vs Revenue")
plt.show()
```

```python
# Line Plot

df.groupby('Year')['Gross Revenue'].mean().plot(kind='line')
plt.title("Average Revenue Over Years")
plt.xlabel("Year")
plt.ylabel("Average Revenue")
plt.show()
```
