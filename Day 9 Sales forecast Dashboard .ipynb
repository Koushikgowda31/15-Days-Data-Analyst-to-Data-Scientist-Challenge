```python
import pandas as pd

# Load the dataset
df = pd.read_csv("Global Superstore 2018.csv",encoding='ISO-8859-1')

# Basic info
print("Shape of dataset:", df.shape)
print("\nColumn names:\n", df.columns)
print("\nData types:\n", df.dtypes)
print("\nMissing values:\n", df.isnull().sum())

# Preview
df.head()
```

```python
# Check column names and data types
df.info()

# Display a few rows
df.head()
```

```python
df.isnull().sum()
```

```python
# Convert 'Order Date' to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True, errors='coerce')

# Sort the data by date
df = df.sort_values('Order Date')
```

```python
# Daily sales
daily_sales = df.groupby('Order Date')['Sales'].sum().reset_index()

# Monthly sales
monthly_sales = daily_sales.resample('M', on='Order Date').sum()

# Plot monthly sales
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 5))
plt.plot(monthly_sales.index, monthly_sales['Sales'], marker='o')
plt.title("Monthly Sales Over Time")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Create a copy to work with
df_ts = df.copy()

# Extract year-month for time series grouping
df_ts['Order Month'] = df_ts['Order Date'].dt.to_period('M')

# Group by month and sum the sales
monthly_sales = df_ts.groupby('Order Month')['Sales'].sum().reset_index()

# Convert 'Order Month' to datetime for plotting and modeling
monthly_sales['Order Month'] = monthly_sales['Order Month'].dt.to_timestamp()

# Preview
print(monthly_sales.head())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(monthly_sales['Order Month'], monthly_sales['Sales'], marker='o')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

# Set the index as datetime for decomposition
monthly_sales.set_index('Order Month', inplace=True)

# Perform decomposition
decomposition = seasonal_decompose(monthly_sales['Sales'], model='additive', period=12)

# Plot decomposition
decomposition.plot()
plt.tight_layout()
plt.show()
```

```python
# Define train-test split (e.g., last 6 months as test)
train = monthly_sales[:-6]
test = monthly_sales[-6:]

print("Train shape:", train.shape)
print("Test shape:", test.shape)
```

```python
# Naive forecast - repeat last value
naive_forecast = test.copy()
naive_forecast['Naive_Prediction'] = train['Sales'].iloc[-1]

# Plot actual vs naive
plt.figure(figsize=(12, 6))
plt.plot(train.index, train['Sales'], label='Train')
plt.plot(test.index, test['Sales'], label='Test')
plt.plot(naive_forecast.index, naive_forecast['Naive_Prediction'], label='Naive Forecast', linestyle='--')
plt.title("Naive Forecast vs Actual Sales")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

mae = mean_absolute_error(test['Sales'], naive_forecast['Naive_Prediction'])
rmse = np.sqrt(mean_squared_error(test['Sales'], naive_forecast['Naive_Prediction']))

print("Naive Forecast MAE:", round(mae, 2))
print("Naive Forecast RMSE:", round(rmse, 2))
```

```python
from statsmodels.tsa.holtwinters import SimpleExpSmoothing

# Fit model on training data
model_ets = SimpleExpSmoothing(train['Sales']).fit()

# Predict on test set
ets_pred = model_ets.forecast(len(test))

# Plot predictions
plt.figure(figsize=(12, 6))
plt.plot(train.index, train['Sales'], label='Train')
plt.plot(test.index, test['Sales'], label='Test')
plt.plot(test.index, ets_pred, label='ETS Forecast', linestyle='--')
plt.title("Exponential Smoothing Forecast vs Actual Sales")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.legend()
plt.grid(True)
plt.show()
```

```python
mae_ets = mean_absolute_error(test['Sales'], ets_pred)
rmse_ets = np.sqrt(mean_squared_error(test['Sales'], ets_pred))

print("ETS Forecast MAE:", round(mae_ets, 2))
print("ETS Forecast RMSE:", round(rmse_ets, 2))
```

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing

# Fit Holt-Winters Model (Additive seasonality & trend)
model_hw = ExponentialSmoothing(
    train['Sales'], 
    trend='add', 
    seasonal='add', 
    seasonal_periods=12
).fit()

# Forecast
hw_pred = model_hw.forecast(len(test))

# Plot
plt.figure(figsize=(12, 6))
plt.plot(train.index, train['Sales'], label='Train')
plt.plot(test.index, test['Sales'], label='Test')
plt.plot(test.index, hw_pred, label='Holt-Winters Forecast', linestyle='--')
plt.title("Holt-Winters Forecast vs Actual Sales")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.legend()
plt.grid(True)
plt.show()
```

```python
mae_hw = mean_absolute_error(test['Sales'], hw_pred)
rmse_hw = np.sqrt(mean_squared_error(test['Sales'], hw_pred))

print("Holt-Winters MAE:", round(mae_hw, 2))
print("Holt-Winters RMSE:", round(rmse_hw, 2))
```

```python
def get_product_timeseries(df, product_name):
    product_df = df[df['Product Name'] == product_name].copy()
    product_df['Month'] = product_df['Order Date'].dt.to_period('M').dt.to_timestamp()
    ts = product_df.groupby('Month')['Sales'].sum()
    ts = ts.asfreq('MS').fillna(0)
    return ts
```

```python
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt

def forecast_product_sales(ts, product_name, periods=6):
    # Fit ARIMA model
    model = ARIMA(ts, order=(1,1,1))
    model_fit = model.fit()

    # Forecast
    forecast = model_fit.forecast(steps=periods)

    # Plot
    plt.figure(figsize=(10, 4))
    plt.plot(ts, label='Historical Sales')
    plt.plot(forecast.index, forecast, label='Forecast', color='orange')
    plt.title(f'{product_name} - Sales Forecast')
    plt.xlabel('Month')
    plt.ylabel('Sales')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
# Core Products
for product in core_top5['Product Name']:
    ts = get_product_timeseries(df, product)
    forecast_product_sales(ts, product)

# Underperforming Products
for product in under_top5['Product Name']:
    ts = get_product_timeseries(df, product)
    forecast_product_sales(ts, product)
```
